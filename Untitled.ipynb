```python
def sine_2(X, signal_freq=100.):
    return (np.sin(2 * np.pi * (X) / signal_freq) + np.sin(4 * np.pi * (X) / signal_freq)) / 2.0

def sine_1(X, signal_freq=100.):
    return np.sin(2 * np.pi * (X) / signal_freq)

def noisy(Y, noise_range=(-0.05, 0.05)):
    noise = np.random.uniform(noise_range[0], noise_range[1], size=Y.shape)
    return Y + noise

def sample(sample_size):
    random_offset = np.random.randint(0, sample_size)
    X = np.arange(sample_size)
    Y = noisy(sine_2(X + random_offset))
    return Y
```

```python
class LSTM(nn.Module):
    def __init__(self, seq_size, hidden_size, out_size):
        # 
        # :param seq_size: 
        # :param hidden_size: 
        # :param out_size: 
        super(LSTM, self).__init__()
        self.xh = torch.nn.LSTM(seq_size, hidden_size, batch_first=True)
        self.hy = torch.nn.Linear(hidden_size, out_size)
        self.hidden_size = hidden_size
#         print(self.hidden_size)
        
    def __call__(self, xs):
        # 
        # :param xs: 
        h, self.hidden = self.xh(xs, self.hidden)
        y = self.hy(h)
#         print(y.shape)
        return y

    def reset(self):
        # 
        self.hidden = (Variable(torch.zeros(1, 1, self.hidden_size)), Variable(torch.zeros(1, 1, self.hidden_size))) # h, c
```

```python
EPOCH_NUM = 300
HIDDEN_SIZE = 5
BATCH_ROW_SIZE = 100 # 
BATCH_COL_SIZE = 20 # 
 
# 
train_data = np.array([np.sin(i*2*np.pi/50) for i in range(50)]*10)
 
# 
train_x, train_t = [], []
for i in range(0, len(train_data)-BATCH_COL_SIZE):
    train_x.append(train_data[i:i+BATCH_COL_SIZE])
#     train_t.append(train_data[i+1:1+i+BATCH_COL_SIZE])
    train_t.append(train_data[i+BATCH_COL_SIZE])
    
train_x = np.array(train_x, dtype="float32")
train_t = np.array(train_t, dtype="float32")
N = len(train_x)

model = RNN(seq_size=BATCH_COL_SIZE, hidden_size=HIDDEN_SIZE, out_size=1)
# model = RNN(input_size=1, hidden_size=HIDDEN_SIZE, num_layers=1, output_size=1)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters())

# 
print("Train")
st = datetime.datetime.now()
for epoch in range(EPOCH_NUM):
    # 
    #  BATCH_ROW_SIZE
    total_loss = 0
    
    for j in range(5):
        x, t = [], []
        
        for i in range(BATCH_ROW_SIZE):
            index = np.random.randint(0, N) # 
            x.append(train_x[index]) # BATCH_COL_SIZE
            t.append(train_t[index])
    #     print(np.array(x).shape)
        x = np.array(x, dtype="float32").reshape(1, 100, 20)
        t = np.array(t, dtype="float32")
        x = Variable(torch.from_numpy(x))
        t = Variable(torch.from_numpy(t))
#         model.reset() # 
        y = model(x)
        loss = criterion(y, t)
        loss.backward()
        total_loss += loss.data[0]
        optimizer.step()
        #     print(total_loss)
    if (epoch+1) % 100 == 0:
        ed = datetime.datetime.now()
        print("epoch:\t{}\ttotal loss:\t{}\ttime:\t{}".format(epoch+1, total_loss, ed-st))
#         print("epoch:\t{}\ttotal loss:\t{}\ttime:\t{}".format(epoch+1, loss.data[0], ed-st))
        
        st = datetime.datetime.now()
```

```python
print("\nPredict")
predict = np.empty(0) # 
inseq_size = 20
inseq = train_data[:inseq_size] # 
for _ in range(N - inseq_size):
#     model.reset() # 
    x = np.array([inseq], dtype="float32").reshape(1, 1, 20)
#     print(x.shape)
    x = Variable(torch.from_numpy(x))
    y = model(x)
    y = y.data.numpy().reshape(1)[0]
    predict = np.append(predict, y)
    inseq = np.delete(inseq, 0)
    inseq = np.append(inseq, y)
 
plt.plot(range(len(train_data)), train_data, color="red", label="t")
plt.plot(range(inseq_size+1, N+1), predict, color="blue", label="y")
plt.legend(loc="upper left")
plt.show()
 
```
